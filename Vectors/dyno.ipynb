```python
#Adds vector b to vector a
def add_v(a, b):
    return tuple((a[0] + b[0], a[1] + b[1]))
```

```python
#Adds vector b to each vector in a
def add(a, b):
    return [add_v(a_v, b) for a_v in a]
```

```python
dino_vectors_2 = add(dino_vectors, b=(1, 0))
```

```python
draw(Points(*dino_vectors), Polygon(*dino_vectors), Points(*dino_vectors_2), Polygon(*dino_vectors_2))
```

```python
#Multiply one vector to another
def mul_v(a, b):
    return (a[0] * b[0], a[1] * b[1])
```

```python
#Multiply each vector in a to b
def mul(a, b):
    return [mul_v(a_v, b) for a_v in a]
```

```python
dino_vectors_3 = add(mul(dino_vectors, b=(0.1, 0.1)), b=(5, 5))
```

```python
draw(Points(*dino_vectors), Polygon(*dino_vectors), Polygon(*dino_vectors_3))
```

```python
def size(vectors, dim):
    positions = [v[dim] for v in vectors]
    return abs(max(positions) - min(positions))
```

```python
def scale(scalar,v):
    return (scalar * v[0], scalar * v[1])
```

```python
def add(*vectors):
    return (sum([v[0] for v in vectors]), sum([v[1] for v in vectors]))
```
