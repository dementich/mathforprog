```python
def pol2cart(distance, angle):
    x = distance * cos(angle)
    y = distance * sin(angle)
    return (x, y)
```

```python
def cart2pol(x, y):
    angle = atan2(y, x)
    distance = sqrt(x**2 + y**2)
    return (distance, angle)
```

```python
def rotate_v(v, r_angle):
    (distance, angle) = cart2pol(v[0], v[1])
    return pol2cart(distance, angle + r_angle)
```

```python
def rotate(v, r_angle):
    return list([rotate_v(v_i, r_angle) for v_i in v])
```

```python
def translate_v(v, tr):
    return (v[0] + tr[0], v[1] + tr[1])
```

```python
def translate(v, tr):
    return [translate_v(v_i, tr) for v_i in v]
```

```python
def regular_polygon(angle_count, radius=1):
    angle = 2 * pi / angle_count
    return [pol2cart(radius, (angle_i + 1) * angle) for angle_i in list(range(angle_count))]
```

```python
draw(Polygon(*regular_polygon(3, 3)))
```
