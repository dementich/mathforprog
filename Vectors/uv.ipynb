# Отдельный блокнот для задачи по uv

**Описание задачи:**

Даны векторы $u = (-1,1)$ и $v = (1,1)$ и действительные числа $r$ и $s$, где $r \in (-1, 1)$, а $s \in (-3, 3)$. Требуется отобразить точки на двумерной плоскости, попадающие в границы выражения $r \cdot u + s \cdot v$

**Решаем**

```python
from vector_drawing import *
import numpy as np
```

```python
u = (-1, 1)
v = (1, 1)
r = np.random.uniform(-1, 1, 500)
s = np.random.uniform(-3, 3, 500)
```

Сначала научимся считать $r \cdot u$ и $s \cdot v$:

```python
def mul(vector, scalar):
    return (vector[0] * scalar, vector[1] * scalar)
```

```python
ru = [mul(u, r_i) for r_i in r]
sv = [mul(v, s_i) for s_i in s]
```

А теперь научимся складывать наборы векторов:

```python
def add(set1, set2):
    #return tuple(sum(v[0] for v in vectors), sum(v[1] for v in vectors))
    return [(ru_i[0] + sv_i[0], ru_i[1] + sv_i[1]) for (ru_i, sv_i) in zip(set1, set2)]
```

```python
draw(Points(*add(sv, ru)))
```
